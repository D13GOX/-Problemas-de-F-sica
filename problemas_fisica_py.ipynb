```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_problemas_fisica():
    # Constante de gravedad
    g = -9.8  # m/s^2
    
    # Problema 1
    print("\nProblema 1:")
    tasa_aceleracion = float(input("Ingrese la tasa de aceleración (en m/s^2): "))
    tiempo = float(input("Ingrese el tiempo (en segundos): "))
    velocidad_final = tasa_aceleracion * tiempo
    desplazamiento = 0.5 * tasa_aceleracion * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

    # Problema 2
    print("\nProblema 2:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo = float(input("Ingrese el tiempo (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo
    desplazamiento = velocidad_inicial * tiempo + 0.5 * aceleracion * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

    # Problema 3
    print("\nProblema 3:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo = float(input("Ingrese el tiempo (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo
    desplazamiento = velocidad_inicial * tiempo + 0.5 * aceleracion * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

    # Problema 4
    print("\nProblema 4:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo = float(input("Ingrese el tiempo (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo
    desplazamiento = velocidad_inicial * tiempo + 0.5 * aceleracion * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

    # Problema 5
    print("\nProblema 5:")
    altura = float(input("Ingrese la altura desde la que cae el objeto (en m): "))
    velocidad_final = np.sqrt(2 * abs(g) * altura)
    tiempo = np.sqrt(2 * altura / abs(g))
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El tiempo de caída es: ", tiempo, "s")

    # Problema 6
    print("\nProblema 6:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    tiempo1 = float(input("Ingrese el tiempo de movimiento a velocidad constante (en segundos): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo2 = float(input("Ingrese el tiempo de aceleración (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo2
    desplazamiento = velocidad_inicial * (tiempo1 + tiempo2) + 0.5 * aceleracion * tiempo2**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento total es: ", desplazamiento, "m")

    # Problema 7
    print("\nProblema 7:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    tiempo = velocidad_inicial / abs(g)
    altura_maxima = velocidad_inicial * tiempo + 0.5 * g * tiempo**2
    print("La altura máxima alcanzada es: ", altura_maxima, "m")
    print("El tiempo que tarda en alcanzarla es: ", tiempo, "s")

    # Problema 8
    print("\nProblema 8:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo = float(input("Ingrese el tiempo (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo
    desplazamiento = velocidad_inicial * tiempo + 0.5 * aceleracion * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

    # Problema 9
    print("\nProblema 9:")
    velocidad_inicial = float(input("Ingrese la velocidad inicial (en m/s): "))
    tiempo1 = float(input("Ingrese el tiempo de movimiento a velocidad constante (en segundos): "))
    aceleracion = float(input("Ingrese la aceleración (en m/s^2): "))
    tiempo2 = float(input("Ingrese el tiempo de desaceleración (en segundos): "))
    velocidad_final = velocidad_inicial + aceleracion * tiempo2
    desplazamiento = velocidad_inicial * (tiempo1 + tiempo2) + 0.5 * aceleracion * tiempo2**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento total es: ", desplazamiento, "m")

    # Problema 10
    print("\nProblema 10:")
    tiempo = float(input("Ingrese el tiempo de caída (en segundos): "))
    velocidad_final = g * tiempo
    desplazamiento = 0.5 * g * tiempo**2
    print("La velocidad final es: ", velocidad_final, "m/s")
    print("El desplazamiento es: ", desplazamiento, "m")

# Llamada a la función para ejecutar los cálculos
calcular_problemas_fisica()
```
